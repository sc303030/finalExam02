```python
import datetime
import pandas as pd
```

```python
rest_day = pd.read_excel('./stock/' + 'stock_rest' + '.xlsx',header=None)
rest_day.iloc[:,0] = rest_day.iloc[:,0].astype(str)
rest_day_list = [i for i in rest_day.iloc[:,0]]
rest_day_list
```

```python
def getDayName(date):
        dayString =  ['월', '화', '수', '목', '금', '토', '일']
        return dayString[datetime.date(int(date[0:4]),int(date[5:7]) ,int(date[-2:])).weekday()]

def day(day_day):
    now = day_day
    Thistime = datetime.datetime.strptime(now, '%Y-%m-%d')
    date_list = []
    for i in range(365 * 8):
        Thistime = Thistime + datetime.timedelta(days=1)
        str_date = Thistime.strftime('%Y-%m-%d')
        date_list.append(str_date)
    date_list_df = pd.DataFrame({
        '일자' : date_list,
    })

    date_list_df['요일'] = [  getDayName(date) for date in date_list_df['일자']]
    date_list_df['일자'] = date_list_df.iloc[:,0].astype(str)                                                                  
    date_list_df['요일'] = date_list_df.iloc[:,1].astype(str)
    return  date_list_df                                                            
```

```python
def dayday(data_name,date_list,rest_day_list):
    data = pd.read_excel('./stock/' + data_name + '.xlsx')
    data = data.iloc[:,:2].copy()
    
    
    data['일자'] = data.iloc[:,0].astype(str)
    data['요일'] = [  getDayName(date) for date in data['일자']]
    data['요일'] = data.iloc[:,2].astype(str)
    
    
    data_merge_df = pd.merge(date_list, data,how='outer')
    
    index_list = data_merge_df[ data_merge_df['요일'] == '일'].index
    data_merge_df.drop(index_list,inplace=True)
    
    index_list2 = data_merge_df[ data_merge_df['일자'] >= '2020-11-01'].index
    data_merge_df.drop(index_list2,inplace=True)
    
    
    for idx, data in enumerate(data_merge_df['일자']):
        if data in rest_day_list:
            data_merge_df.iloc[idx,2] = data_merge_df.iloc[idx-1,2]
            
            
    for idx, date in enumerate(data_merge_df['요일']):
        if date == '토':
            data_merge_df.iloc[idx,2] = data_merge_df.iloc[idx-1,2]
            
            
    for idx, data in enumerate(data_merge_df['일자']):
        if data in rest_day_list:
            data_merge_df.iloc[idx,2] = data_merge_df.iloc[idx-1,2]
            
            
    for idx, date in enumerate(data_merge_df['요일']):
        if date == '토':
            data_merge_df.iloc[idx,2] = data_merge_df.iloc[idx-1,2]   
            
            
    return data_merge_df
```

### asiatech

```python
date_list_df = day('2013-01-01')
asiatech_df = dayday('asiatech', date_list_df,rest_day_list)
asiatech_df.to_excel('./stock/asiatech.xlsx',encoding='utf-8',index=False)
```
