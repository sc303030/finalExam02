```python
from selenium import webdriver
import time
path = './driver/chromedriver.exe'
driver = webdriver.Chrome(path)
```

```python
import pandas as pd
```

```python
date = pd.read_excel('korean.xlsx',header=None)
date_list = [ i for i in date.iloc[:,0]]
```

```python
def rate_crr(data_list):
    result_rate_list = []
    result_date_list = []
    for date in date_list:
        driver.get('https://www.nielsenkorea.co.kr/tv_terrestrial_day.asp?menu=Tit_1&sub_menu=1_1&area=00&begin_date='+str(date))
        page = driver.find_elements_by_css_selector('.ranking_tb')
        page_tr = page[0].find_elements_by_tag_name('tr')
        name_list = []
        rate_list = []
        for tr in page_tr:
            page_td = tr.find_elements_by_tag_name('td')
            for idx,td_text in enumerate(page_td):
                if len(name_list) == 0:
                    if td_text.text == '한국인의밥상':
                        name_list.append(td_text.text)
                else:
                    if len(rate_list) == 0:
                        rate_list.append(td_text.text)
                        result_date_list.append(date)
                        result_rate_list.append(td_text.text)
                        break  

                        
    return  result_rate_list,  result_date_list


```

```python
result_rate_list,  result_date_list = rate_crr(date_list)
```

```python
print(result_rate_list[:10])
print(result_date_list[:10])
```

```python
rate_csv = pd.DataFrame({
    'date' : result_date_list,
     'ratio'  : result_rate_list
                        })
rate_csv
```

```python
rate_csv.to_csv('./data/rate_csv.csv',mode='w',encoding='utf-8',index=False)
```
